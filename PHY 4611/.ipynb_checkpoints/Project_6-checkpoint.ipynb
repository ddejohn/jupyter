```julia
using Plots
gr()

##
# Main programs

function analytic(n; gp=50, V=1., Lx=1., Ly=1.)
    hx = Lx/(gp-1)
    hy = Ly/(gp-1)
    
    srf = zeros(gp, gp)
    for i in 1:gp
        srf[i,:] = [(V*4/pi)*sum((sin(k*pi*(j-1)*hx/Lx)*sinh(k*pi*(i-1)*hy/Lx))/(k*sinh(k*pi*Ly/Lx))
                        for k in 1:2:n) for j in 1:gp]
    end
    
    return (srf, [x*hx for x in 0:(gp-1)], [y*hy for y in 0:(gp-1)])
end

function laplace(; mthd, gp=50, V=1., L=1.)
    h = L/(gp-1)
    
    vol = zeros(gp, gp, gp)
    # BCs: V(x,y,0) = V(x,y,L) = V0
    vol[:, :, 1] .= V
    vol[:, :, end] .= V
    
    return (mthd(vol), [x*h for x in 0:(gp-1)], [y*h for y in 0:(gp-1)], [z*h for z in 0:(gp-1)]) 
end

##
# Integrators

# TODO: outer loop to iterate until steady state solution reached.

function ftcs(v)
    gp = size(v,1)
    # for _ in ... 
    for k in 2:(gp-1)
        for i in 2:(gp-1)
            for j in 2:(gp-1)
                v[i,j,k] = 1/4*(v[i+1,j,k-1] + v[i-1,j,k-1] + v[i,j+1,k-1] + v[i,j-1,k-1])
            end
        end
    end
    return v
end

function gauss(v)
    gp = size(v,1)
    # for _ in ...
    for k in 2:(gp-1)
        for i in 2:(gp-1)
            for j in 2:(gp-1)
                v[i,j,k] = 1/4*(v[i+1,j,k-1] + v[i-1,j,k] + v[i,j+1,k-1] + v[i,j-1,k])
            end
        end
    end
    return v
end

function sor(v)
    gp = size(v,1)
    # for _ in ...
    w = 2/(1+sin(pi/gp))
    for k in 2:(gp-1)
        for i in 2:(gp-1)
            for j in 2:(gp-1)
                v[i,j,k] = (1-w)*v[i,j,k-1] + w/4*(v[i+1,j,k-1] + v[i-1,j,k] + v[i,j+1,k-1] + v[i,j-1,k])
            end
        end
    end
    return v
end

##
# Plotting

function srf_plot(dat)
    srf, xx, yy = dat
    
    p1=surface(xx, yy, srf, title="",
                xlabel=:"x", ylabel=:"y", zlabel=:"V(x,y)", lims=:round,
                xticks=0:0.5:1., yticks=0:0.5:1.,
                color=:Spectral_r, colorbar=false,
                st=:surface, camera=(45,45))

    p2=contour(xx, yy, srf, title="",
                xlabel=:"x", ylabel=:"y", 
                xticks=0:0.5:1., yticks=0:0.5:1.,
                levels=[i for i in 0.:0.1:1.],
                color=:Spectral_r, colorbar=false,
                st=:contourf, contour_labels=true)

    plot(p1, p2, layout=2, size=(950,475))
end

# function vol_gif(dat)
#     vol, xx, yy, zslices = dat

#     anim = @animate for i in 1:size(vol,1)
#         zz = vol[:,:,i]
#         zi = "$(round(zslices[i], digits=2))"
#         if length(zi) < 5
#             zi = rpad(zi, 5, "0")
#         end
#         ttl = string("Potential inside a cube at z = ", zi)
#         p1=surface(xx, yy, zz, title=ttl,
#                     xlabel=:"x", ylabel=:"y", zlabel=:"V(x,y)", lims=:round,
#                     xticks=0:0.5:1., yticks=0:0.5:1.,
#                     color=:Spectral_r, colorbar=false,
#                     st=:surface, camera=(45,45))

#         p2=contour(xx, yy, zz, title="",
#                     xlabel=:"x", ylabel=:"y", 
#                     xticks=0:0.5:1., yticks=0:0.5:1.,
#                     color=:Spectral_r, colorbar=false,
#                     st=:contourf, contour_labels=true)

#         plot(p1, p2, layout=2, size=(950,475))
#     end every 1
    
#     gif(anim, "test.gif")
# end

function one_percent()
    nums=[11, 21, 51, 151]
    gp = 1000
    lft = Int(round(gp*0.4))
    rt = gp-lft+1
    
    hz = [i/gp for i in (lft-1):rt]
    v0 = [1. for i in (lft-1):rt]
    vp = [1.005 for i in (lft-1):rt]
    vm = [0.995 for i in (lft-1):rt]

    p = plot(title="Partial sum to n terms on $gp grid points",
             size=(900,900), yticks=0.9:0.01:1.1, legend=:bottomright)
    
    plot!(hz, v0, l=(3,:dash,:green), alpha=0.35, label="V_0")
    plot!(hz, vp, l=(3,:dash,:red), alpha=0.35, label="+0.5%")
    plot!(hz, vm, l=(3,:dash,:blue), alpha=0.35, label="-0.5%")

    for n in nums
        dat = analytic(n, gp=1000)
        srf, xx, yy = dat
        plot!(p, xx[lft:rt], srf[end, lft:rt], lw=3, label="n=$n")
    end

    p
end;
```

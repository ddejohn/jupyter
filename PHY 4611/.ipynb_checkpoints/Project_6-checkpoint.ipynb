```julia
using Plots
gr()

function analytic(n; gp=50, V=1., Lx=1., Ly=1.)
    hx = Lx/(gp-1)
    hy = Ly/(gp-1)
    
    srf = zeros(gp, gp)
    for i in 1:gp
        srf[i,:] = [(V*4/pi)*sum((sin(k*pi*(j-1)*hx/Lx)*sinh(k*pi*(i-1)*hy/Lx))/(k*sinh(k*pi*Ly/Lx))
                        for k in 1:2:n) for j in 1:gp]
    end
    
    return (srf, [x*hx for x in 0:(gp-1)], [y*hy for y in 0:(gp-1)])
end

function potential(; mthd, gp=50, V=1., Lx=1., Ly=1.)
    hx = Lx/(gp-1)
    hy = Ly/(gp-1)
    
    srf = zeros(gp, gp)
    srf[1, 2:end-1] .= 1.
    
    # Put neighboring points into mthd
    for i in 2:(gp-1)
        for j in 2:(gp-1)
            # point-left, point-right, -up, -down
            # pl, pr, pu, pd, pf, pb = srf[i,j-1], srf[i,j+1], srf[i+1,j], srf[i-1,j]
            # srf[i,j] = mthd(pl, pr, pu, pd, pf, pb)
        end
    end
    
    return (srf, [x*hx for x in 0:(gp-1)], [y*hy for y in 0:(gp-1)])
end

function srf_plot(dat)
    srf, xx, yy = dat
    
    p1=surface(xx, yy, srf, title="",
                xlabel=:"x", ylabel=:"y", zlabel=:"V(x,y)", lims=:round,
                xticks=0:0.5:1., yticks=0:0.5:1.,
                color=:Spectral_r, colorbar=false,
                st=:surface, camera=(45,45))

    p2=contour(xx, yy, srf, title="",
                xlabel=:"x", ylabel=:"y", 
                xticks=0:0.5:1., yticks=0:0.5:1.,
                levels=[i for i in 0.:0.1:1.],
                color=:Spectral_r, colorbar=false,
                st=:contourf, contour_labels=true)

    plot(p1, p2, layout=2, size=(950,475))
end;
```

```julia
srf_plot(analytic(11))
```

```julia
srf_plot(analytic(21))
```

```julia
srf_plot(analytic(51))
```

```julia
nums=[11, 21, 51, 151]
gp = 1000
left = Int(round(gp*0.4))
right = gp-left+1

p = plot(title="Partial sum to n terms on $gp grid points", size=(900,900), yticks=0.9:0.01:1.1, legend=:bottomright)
plot!([i/gp for i in (left-1):right], [1. for i in (left-1):right], l=(3,:dash,:green), alpha=0.35, label="V_0")
plot!([i/gp for i in (left-1):right], [1.005 for i in (left-1):right], l=(3,:dash,:red), alpha=0.35, label="+0.5%")
plot!([i/gp for i in (left-1):right], [0.995 for i in (left-1):right], l=(3,:dash,:blue), alpha=0.35, label="-0.5%")

for n in nums
    dat = analytic(n, gp=1000)
    srf, xx, yy = dat
    plot!(p, xx[left:right], srf[end, left:right], lw=3, label="n=$n")
end

p
```

```julia
using Plots
gr()

#——————————————————————————————————————————————————————————————————————————————————————————————————#
# Constants

# const μ = 1.3271244004193938e20 # JPL ephermerides, August 2018
const μ = 4*pi^2
const AU = 1.495978707e11 # JPL ephermerides, August 2018
const m_h = 2.2e14 # Halley's comet mass (wikipedia)

#——————————————————————————————————————————————————————————————————————————————————————————————————#
# Orbital state vector

struct State
    r::AbstractVector{Float64}
    v::AbstractVector{Float64}
end

Base.:*(k::Number, S::State) = State(k*S.r, k*S.v)
Base.:+(A::State, B::State) = State(A.r + B.r, A.v + B.v)

#——————————————————————————————————————————————————————————————————————————————————————————————————#
# Methods

# Already mass-normalized (little m divided out)
function cent_acc(r::AbstractVector{Float64})
    return -(μ/nrm(r)^3) * r
end

function nrm(r)
    return sqrt(sum(r.^2))
end

function dX(X::State)
    dr = X.v
    dv = cent_acc(X.r)
    return State(dr, dv)
end

function sX(X::State)
    return X + τ/2 * dX(X)
end

function euler(X::State, τ::Float64)
    return X + τ * dX(X)
end

function rk(X::State, τ::Float64)
    return X + τ * dX(sX)
end

function rk4!(X::State, τ::Float64)
    k1 = dX(X)
    k2 = dX(X + τ/2*k1)
    k3 = dX(X + τ/2*k2)
    k4 = dX(X + τ*k3)

    return X + τ/6 * (k1 + 2*k2 + 2*k3 + k4)
end

function test_plot()
    s = State([1, 0], [0, pi])

    ρ = []
    θ = []

    for i in 1:200
        push!(ρ, nrm(s.r))
        push!(θ, mod(atan(s.r[2], s.r[1]), 2pi))
        s = rk4!(s, 0.005)
    end

    p = plot([0], [0], label=:"", proj=:polar, size=(900,900))
    plot!(p, θ, ρ, label=:"")
    p
end;
```

```julia
test_plot()
```

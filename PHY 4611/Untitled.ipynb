```julia
using Plots
# plotlyjs(fmt=:svg)
gr()

const L = 1. # System size
const c = 1. # wave speed

function advect(t, n; mthd)
    # Grid spacing.
    h = L/n

    k = c*t/(2*h) # Coefficient for all schemes.
    klw = 2*k^2 # Coefficient for Lax-Wendroff.

    # Initial conditions; integrate to t=1.0 second.
    steps = Int(round(1/t))
    
    sig = 0.1 # Width of Gaussian pulse.
    wn = pi/sig # Wave number.

    # x- and time-values for plotting.
    xvals = [h*i - L/2 for i in 0:(n-1)]
    tvals = [t*i for i in 0:(steps-1)]
    
    # Initial wv pulse.
    wv = zeros(steps, n)
    for (i,x) in enumerate(xvals)
        wv[1,i] = exp(-x^2/(2*sig^2)) * cos(wn*x)
    end

    return (mthd(wv, n, steps, k, klw), xvals, tvals, n, t)
end

function ftcs(x...)
    wv, n, steps, k = x

    for i in 2:steps
        wv[i,1] = wv[i-1,1] - k*(wv[i-1,n] - wv[i-1,2])
        for j in 2:(n-1)
            wv[i,j] = wv[i-1,j] - k*(wv[i-1,j-1] - wv[i-1,j+1])
        end
        wv[i,n] = wv[i-1,n] - k*(wv[i-1,n-1] - wv[i-1,1])
    end
    
    return wv
end

function lax(x...)
    wv, n, steps, k = x

    for i in 2:steps
        
        wv[i,1] = 0.5*(wv[i-1,2] + wv[i-1,n]) - k*(wv[i-1,2] - wv[i-1,n])
        wv[i,n] = 0.5*(wv[i-1,1] + wv[i-1,n-1]) - k*(wv[i-1,1] - wv[i-1,n-1])
        
        for j in 2:(n-1)
            wv[i,j] = 0.5*(wv[i-1,j+1] + wv[i-1,j-1]) - k*(wv[i-1,j+1] - wv[i-1,j-1])
        end
    end
    
    return wv
end

function wdf(x...)
    wv, n, steps, k, klw = x

    for i in 2:steps
        wv[i,1] = wv[i-1,1] - k*(wv[i-1,2] - wv[i-1,n]) + klw*(wv[i-1,2] + wv[i-1,n] - 2*wv[i-1,1])
        
        for j in 2:(n-1)
            wv[i,j] = wv[i-1,j] - k*(wv[i-1,j-1] - wv[i-1,j+1]) + klw*(wv[i-1,j-1] + wv[i-1,j+1] - 2*wv[i-1,j])
        end
        
        wv[i,n] = wv[i-1,n] - k*(wv[i-1,1] - wv[i-1,n-1]) + klw*(wv[i-1,1] + wv[i-1,n-1] - 2*wv[i-1,n])
    end
    
    return wv
end

function plt_dat(dat)
    wv, x, t, nstep, tstep = dat
    println(nstep)
    ttl = string("Advection for dt = ", string(tstep), " s, N = ", string(nstep))
    p1=surface(x, t, wv, title=ttl,
            xlabel=:"x", ylabel=:"time", zlabel=:"a(x,t)", 
            color=:Spectral_r, colorbar=false,
            st=:surface, camera=(35,60))
    p2=plot(x, wv[1,:], label="Initial")
    plot!(p2, x, wv[end,:], label="Final")
    plot(p1, p2, layout=2, size=(950,450))
end;
```

```julia
using Plots
gr()

# UGLY AF: FIX ME
function analytic3d(; n, z, gp=50, V=1., L=1.)
    h = L/(gp-1)    
    vol = zeros(gp,gp)
    
    # Fourier double sum
    for p in 1:2:n
        for q in 1:2:n
            for i in 1:gp
                for j in 1:gp
                    cnm = 16*V*sech(pi*sqrt(p^2 + q^2)/2)/(p*q*pi^2)
                    xx = sin(p*pi*(i-1)*h/L)
                    yy = sin(q*pi*(j-1)*h/L)
                    zz = cosh(pi*sqrt(p^2 + q^2)*(z - L/2)/L)
                    vol[i,j] += cnm*xx*yy*zz
                end
            end
        end
    end
    
    return (vol, [x for x in 0:h:L], [y for y in 0:h:L])
end
    
# Calculate potential in a cube.
# mthd = integrator, gp = # of grid points, int = interior values function.
function laplace(; mthd, gp=50, V=1., L=1., intr=zeroint)
    h = L/(gp-1)
    
    # Initialize our cube with points defined by an intr function. Defaults to all zeros.
    vol = intr(gp)
    
    # Boundary conditions (the other boundary conditions are satisfied by intr).
    vol[:, :, 1] .= V
    vol[:, :, end] .= V
    
    # For plotting.
    xx = [x for x in 0:h:L]
    yy = [y for y in 0:h:L]
    zz = [z for z in 0:h:L]
    
    # Calculate the potential using given mthd.
    v, imax = mthd(vol)
    return (v, xx, yy, zz, uppercase(string(mthd)), imax)
end

#——————————————————————————————————————————————————————————————————————————————————————————————————————————————————————#
# Helper functions

# Zeros on the interior points.
zeroint(gp) = zeros(gp,gp,gp)

# Interior points randomized.
function randint(gp)
    vol = rand(gp,gp,gp)
    for z in 2:(gp-1)
        vol[1,:,z] .= 0.
        vol[:,1,z] .= 0.
        vol[end,:,z] .= 0.
        vol[:,end,z] .= 0.
    end
    return vol
end

# Interior points given by first 5 terms of analytic soln.
function analyticint(gp)
    vol = zeros(gp,gp,gp)
    for i in 2:(gp-1)
        zz, xx, yy = analytic3d(n=5, z=(i/gp), gp=gp)
        vol[:,:,i] = zz
        vol[1,:,i] .= 0.
        vol[:,1,i] .= 0.
        vol[end,:,i] .= 0.
        vol[:,end,i] .= 0.
    end
    return vol
end

#——————————————————————————————————————————————————————————————————————————————————————————————————————————————————————#
# Integrators

function sor1(v)
    gp = size(v,1)
    w = 2/(1+sin(pi/gp))
    div = (gp-2)^2
    imax = 0
    
    for iter in 1:gp^2
        c_sum = 0
        
        for k in 2:(gp-1)
            for i in 2:(gp-1)
                for j in 2:(gp-1)
                    temp = w/6*(v[i+1,j,k] + v[i-1,j,k] + 
                                v[i,j+1,k] + v[i,j-1,k] + 
                                v[i,j,k+1] + v[i,j,k-1]) + (1-w)*v[i,j,k]
                    
                    c_sum += abs(1-v[i,j,k]/temp)
                    v[i,j,k] = temp
                end
            end
        end
        
        imax = iter
        if c_sum/div < 1e-4
            break
        end
    end
    
    return (v,imax)
end

function sor2(v)
    gp = size(v,1)
    w = 2/(1+sin(pi/gp))
    div = (gp-2)^2
    imax = 0
    
    # Matrix slices
    L = 1:gp-2      # Left
    M = 2:gp-1      # Middle
    R = 3:gp        # Right
    
    for iter in 1:gp^2
        c_sum = 0.
        imax = iter
        
        temp = (1-w)*v[M,M,M] + w/6*(v[L,M,M] + v[R,M,M] + 
                                     v[M,L,M] + v[M,R,M] + 
                                     v[M,M,L] + v[M,M,R])
        
        c_sum = sum(@. abs(1-v[M,M,M]/temp))
        v[M,M,M] = temp
        
        if c_sum/div < 1e-4
            break
        end
    end
    
    return (v,imax)
end

#——————————————————————————————————————————————————————————————————————————————————————————————————————————————————————#
# Plotting

function srf_plot(dat; ttl="", zslc="", zlm=(0.,1.))
    zz, xx, yy = dat
    
    st = string(ttl, "\n", "\n V(x, y", zslc, ")")    
    srf=surface(xx, yy, zz, title=st,
                xlabel="x", ylabel="y", zlabel="",
                xticks=0:0.5:1., yticks=0:0.5:1.,
                zlims=zlm,
                color=:Spectral_r, colorbar=false,
                st=:surface, camera=(45,45))

    con=contour(xx, yy, zz, title=string("Equipotential level curves"),
                xlabel="x", ylabel="y", 
                xticks=0:0.25:1., yticks=0:0.25:1.,
                levels=[i for i in 0.:0.1:1.],
                color=:Spectral_r, colorbar=false,
                st=:contourf, contour_labels=true)
    
    plot(srf, con, size=(950,475))
end

function vol_plot(dat; lvl)
    vol, xx, yy, zslices, mthd, imax = dat
    gp = size(vol,3)
    zz = vol[:,:,Int(round(gp*lvl)+1)]
    ttl = "$(mthd), n = $(imax+1) iterations"
    srf_plot((zz,xx,yy), ttl=ttl, zslc=", z = $(lvl)")
end;
```

```julia
vol_plot(laplace(mthd=sor2), lvl=0.5)
```

```julia
vol_plot(laplace(mthd=sor1), lvl=0.025)
```

```julia
function vol_gif(dat)
    vol, xx, yy, zslices, mthd, imax = dat
    gp = size(vol,3)

    anim = @animate for i in 2:(size(vol,1)-1)
        zz = vol[:,:,i]
        zi = "$(round(zslices[i], digits=2))"
        if length(zi) < 4
            zi = rpad(zi, 4, "0")
        end
        
        ttl = "Potential in a cube, $(mthd) method, $(imax+1) iterations"
        srf_plot((zz,xx,yy), ttl=ttl, zslc=", z = $zi")
    end every 1
    
    gif(anim, string("laplace_", mthd, ".gif"))
end;
```

#### 1) State a major difference between the nature of roundoff errors and discretization errors.

Roundoff errors occur due to the finite number of bits available to be used to represent a real number, limiting the precision with which the number can be stored in binary. Discretization errors are due to the truncation of infinite series representations of functions, since only a finite number of terms can be realistically used. 

#### 2) Carry out a derivation using a Taylor series to define the approximation

$$
f'(x_0) = \frac{f(x_0 + h) - f(x_0 - h)}{2h}
$$

Starting with the approximations,

$$
    f(x_0 + h) = f(x_0) + h\cdot f'(x_0) + \frac{h^2}{2}f''(x_0) + \frac{h^3}{6}f^{(3)}(x_0) \;+\; \dots \;\;\;\;\;\;\;\;\;\; (1) \\
    f(x_0 - h) = f(x_0) - h\cdot f'(x_0) + \frac{h^2}{2}f''(x_0) - \frac{h^3}{6}f^{(3)}(x_0) \;+\; \dots \;\;\;\;\;\;\;\;\;\; (2) \\
$$

Subtracting $(2)$ from $(1)$, we have

\begin{align*}
f(x_0 + h) - f(x_0 - h) &= 2hf'(x_0) + \frac{h^3}{3}f^{(3)}(x_0) \;+\; \dots \\
\frac{f(x_0 + h) - f(x_0 - h)}{2h} &= f'(x_0) + \frac{h^2}{6}f^{(3)}(x_0) \;+\; \dots \\
f'(x_0) &\approx \frac{f(x_0 + h) - f(x_0 - h)}{2h} - \frac{h^2}{6}f^{(3)}(x_0)
\end{align*}

Finally, we arrive at

$$
f'(x_0) \approx \frac{f(x_0 + h) - f(x_0 - h)}{2h}
$$

Notice that only the odd-numbered derivatives survive the subtraction $(1) - (2)$. After dividing both sides by $2h$, we're left with the odd derivatives and coefficients $h^2, h^4, h^6, \dots$

If we take $h$ to be small, then only the $h^2$ term will contribute in any significant way, and so we truncate the remaining terms of the infinite series in order to define the initial approximation. And so then the error is $\mathcal{O}(h^2)$ if we remove the $-\frac{h^2}{6}f^{(2)}(x_0)$ term to arrive at our final approximation of the derivative of $f$ at $x=x_0$.


#### 3) Comment on how your approximation in the previous problem compares with the example in the text.

The example in the text is what I believe is called a *forward difference* approximation. The approximation derived in the previous problem **averages** the value of the function at $x = x_0 + h$ and $x = x_0 - h$ and is called the *central difference*.

#### 4) Assess the conditioning of the function $g(x)$ near $x = 0$ as the positive parameter $c$ grows where

$$
g(x) = \tanh(cx) = \frac{e^{cx} - e^{-cx}}{e^{cx} + e^{-cx}}
$$

Let $a = 0.01$ and we use a Taylor series to approximate $g'(x)$ near $x = a$.

First,

\begin{align}
    g'(x) &= c - c\cdot\tanh^2(c\,x) \\ 
    &= c\,(1-\tanh^2(c\,x)) \\
    &= c\,\text{sech}^2(c\,x) \\
    &= \frac{c}{\cosh^2(c\,x)}
\end{align}

And then our Taylor series approximation $\left(\mathcal{O}(h^2)\right)$,

$$
    g'(x) \approx g(a) + g'(a)(x-a)
$$

becomes

$$
    g'(x) \approx \tanh(0.01) + \frac{c}{\cosh^2(c\cdot0.01)}(x-0.01)
$$

near $x=0$. Notice that as $a \rightarrow 0 \;,\; \cosh^2(a) \rightarrow 1$ so then $\frac{c}{\cosh^2(c\cdot0.01)} \rightarrow c$.

Let $c = 1000$, then $\frac{1000}{\cosh^2(1000\times0.01)} \approx 8.245\times 10^{-6}$. Now let $a = 0.001$ and we can see that $\frac{1000}{\cosh^2(1000\times0.001)} \approx 4.1997\times 10^{2}$.

This relatively small change from $a = 0.01 \rightarrow a = 0.001$ has resulted in a change in output by eight orders of magnitude!

Clearly, as $c$ grows large $\tanh(c\cdot x)$ becomes ill-conditioned.

The last bit of accuracy we can retain in a 32-bit word is $2^{-23}$. Anything smaller cannot be distinguished in a 23-bit mantissa. The largest of these inaccuracies comes from what would be the 24th bit, $2^{-24} \approx 5.96\times 10^{-8}$.

### Exercise 6

I've done this exercise many times before and am aware of the pitfalls of calculating large Fibonacci numbers naively. I'm using a built-in function wrapper (most modern languages have some kind of memoization) to cut down significantly on the run time.

Memoization looks up values in a hash map or dictionary and returns if the value is found, otherwise performs the calculation and stores the result. Calculating Fibonacci numbers recursively is a pretty standard use-case of memoization as it allows the program to lookup the previously calculated Fibonacci number without needing to recursively sum through the previous numbers, which is extremely computationally expensive to do naively. In fact, recursive Fibonacci number generation without memoization is on the order of $\mathcal{O}(2^n)$ which can be seen easily by analyzing the recursion tree.

Since the exercise calls for generating a list of $n$ fibonacci numbers, I defined two versions of the list generator: `fvec_fib` and `bvec_fib`. One generates the Fibonacci numbers backwards (`bvec_fib`), the other forwards.

Interestingly, there is an appreciable difference in the execution time between the two, as one calculates the $nth$ Fibonacci number first, and then iterates backwards through the remaining numbers. The other calculates up to the $nth$ number.

From what I can intuit, the backwards Fibonacci generator essentially stores all of the values up to the $nth$ in one method call. Those values all are all stored in cache and are readily available when we move to the $(n-1)th$ Fibonacci number.

The other calculates the first number, then the second number by looking up the first, then the third by looking up the second, and so on. To be honest I'm not entirely certain I understand how Julia implements `memoize` under the hood, but the difference in execution time seems to indicate that the memory lookups occur more when iterating forward.

Anyway, I thought it was interesting enough to point out  \\\_()_/

```julia
# memoized fibonacci;
# Look up fib(n) in cache. If not there, calculate and store the result.
@memoize function fib(n)
    # parent functions handle non-negative and non-integer arguments
    # so just return n itself instead of lots of tedious if statements
    if n < 2
        return n
    end
    return fib(n-1) + fib(n-2) # bog-standard recursion
end;


# generate an array of the first n fibonacci numbers;
function fvec_fib(n)
    # throw an exception if the given argument is not a positive integer
    if typeof(n) == Int && n > 0
        out = []
        for i in 1:n
            push!(out, fib(i)) # push! is analogous to list.extend() in Python
        end
        return out
    end
    return "Argument must be a non-negative integer!"
end


# same as fvec_fib except we iterate backwards from the n-th fibonacci number down to 1.
function bvec_fib(n)
    if typeof(n) == Int && n > 0
        out = []
        for i in n:-1:1
            push!(out, fib(i))
        end
        return reverse(out)
    end
    return "Argument must be a non-negative integer!"
end

println("forward fib time: ", @elapsed forward_fib = fvec_fib(42))
println("backward fib time: ", @elapsed backward_fib = bvec_fib(42))
println("42nd Fibonacci number: ", backward_fib[42])
println("error catching demonstration: ", fvec_fib(-151.5))
```

```julia
# Julia is a relatively new language, and as such its plotting capabilities are very limited.
# There are essentially two copies of each plotted function:
# one plots only every 6th element in the original xrange as scatter plot
# with markers to differentiate the different functions the other is an empty 
# plot with the color and marker of the original curve used only for the legend 
function basic_plot()
    xx = collect(-pi:0.01:pi)
    xm = xx[1:trunc(Int, length(xx)/6):end]

    p = plot(
        title = "super duper fun curves!",
        xlabel="x", ylabel="y", ylim=(-10, 10),
        size=(900,900)
    )

    # actual curves plotted from -pi to pi in increments of 0.01
    # ! syntax denotes a mutation to an existing reference
    plot!(p, xx, xx, lw=3, label="x", alpha=0.5)
    plot!(p, xx, sin.(xx), lw=3, label="", alpha=0.5)
    plot!(p, xx, cos.(xx), lw=3, label="", alpha=0.5)
    plot!(p, xx, tan.(xx), lw=3, label="", alpha=0.5)

    scatter!(p, xm, sin.(xm), m=(:star5, 6), label="")
    scatter!(p, xm, cos.(xm), m=(:d, 5), label="")
    scatter!(p, xm, tan.(xm), m=(:rtriangle, 6), label="")

    plot!(p, [], [], m=:star5, label="sin(x)")
    plot!(p, [], [], m=:d, label="cos(x)")
    plot!(p, [], [], m=:rtriangle, label="tan(x)")
    
    p
end


function sub_plotter()
    # convert a range object into an array
    xx = collect(-pi:0.01:pi)
    # initialize a new array using every 6th element of xx
    xm = xx[1:trunc(Int, length(xx)/6):end]

    p1 = plot(
        xx, xx, lw=3, alpha=0.5,
        title="y = x", label="x",
        xlabel="x", ylabel="y", ylim=(-pi, pi)
    )

    p2 = plot(
        xx, sin.(xx), lw=3, alpha=0.5,
        title="y = sin(x)", label="",
        xlabel="x", ylabel="y", ylim=(-pi, pi)
    )
    scatter!(p2, xm, sin.(xm), m=(:star5, 6), label="")
    plot!(p2, [], [], m=:star5, label="sin(x)")

    p3 = plot(
        xx, cos.(xx), lw=3, alpha=0.5,
        title="y = cos(x)", label="",
        xlabel="x", ylabel="y", ylim=(-pi, pi)
    )
    scatter!(p3, xm, cos.(xm), m=(:d, 5), label="")
    plot!(p3, [], [], m=:d, label="cos(x)")

    p4 = plot(
        xx, tan.(xx), lw=3, alpha=0.5,
        title="y = tan(x)", label="",
        xlabel="x", ylabel="y", ylim=(-pi, pi)
    )
    scatter!(p4, xm, tan.(xm), m=(:rtriangle, 6), label="")
    plot!(p4, [], [], m=:rtriangle, label="tan(x)")
    
    plot(p1, p2, p3, p4, layout=4, size=(900,900))
end;
```
